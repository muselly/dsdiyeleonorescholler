```python
#import valence ratings as panda data frame
valence_df = pd.read_csv('./annotations/annotations-averaged-per-song/dynamic(per-second-annotations)/valence.csv',index_col = 0)
#arousal
arousal_df = pd.read_csv('./annotations/annotations-averaged-per-song/dynamic(per-second-annotations)/arousal.csv',index_col = 0)
arousal_desc1 = arousal_df.describe()
arousal_desc2 = arousal_df.transpose().describe().transpose()
```

```python
valence_df.head()
```

```python
valence_desc1 = valence_df.describe()
valence_desc1.head()
```

```python
valence_dft = valence_df.transpose()
valence_desct = valence_dft.describe()
valence_desct.head()
```

```python
#Plotting valence ratings for song 2

plt.plot(valence_df.iloc[0])
plt.xlabel("500ms samples")
plt.ylabel("Valence rating")
plt.xticks(list(range(0,61,20)))
plt.yticks([-1,0,1])
plt.show()
```

```python
valence_desc2 = valence_desct.transpose()
trainSongs_valence_desc = valence_desc2.query('count == 60')
trainSongs_valence_desc.head()

trainSongs_arousal_desc = arousal_desc2.query('count == 60')
```

```python
#make a list of the song_ids that correspond to the 30sec songs 
trainSongs_ids = list([])
for index in range(len(trainSongs_valence_desc)):
    trainSongs_ids.append(trainSongs_valence_desc.iloc[index].name)
    
#and extract them from the original data frame
trainSongs_valence_df = valence_df.loc[trainSongs_ids]
trainSongs_valence_df.head()
trainSongs_arousal_df = arousal_df.loc[trainSongs_ids]
```

```python
plt.hist(trainSongs_valence_desc['mean'], 50)
```

```python
plt.scatter(trainSongs_arousal_desc['mean'], trainSongs_valence_desc['mean'], s = 2, c = 'red')
plt.xlabel('mean arousal ratings per song')
plt.ylabel('mean valence ratings per song')
plt.show()
```

```python
#for song 2: scatterplot of arousal and valence ratings through time
plt.scatter(arousal_df.iloc[2,:1223], valence_df.iloc[2,:1223])
plt.xlabel('arousal rating')
plt.ylabel('valence rating')
plt.show()
```
